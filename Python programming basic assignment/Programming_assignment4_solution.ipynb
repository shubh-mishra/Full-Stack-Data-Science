1. Write a program to find the factorial of a number

```python
def nfact(num):
    result = 1
    for i in range(1, num+1):
        result*= i
        if i == num:
            return result
```

```python
print('Enter a positive integer: ')
num = int(input())
y = nfact(num)
print('Factorial of ', num, ' is: ', y)
```

2. Write a program to display the multiplication table

```python
def nTable(num):
    for i in range(1, 11):
        print(num, ' * ', i, ' = ', num*i)
```

```python
print('Enter a number: ')
num = int(input())
nTable(num)
```

3. Write a program to print the fibonacci series

```python
def fibonacci(num):
    a=0
    b=1
    if a<=num:
        print(a)
    if b<=num:
        print(b)
    while(True):
        c=a+b
        if c<=num:
            print(c)
        else:
            return
        a=b
        b=c
```

```python
print('Enter a number till where a series needs to be printed: ')
num = int(input())
fibonacci(num)
```

4. Write a program to check Armstrong number

```python
def calcArmstrong(num, val):
    result = 0
    for i in range(0, val):
        dig = num%10
        num = int(num/10)
        result += pow(dig,val)
    return result
```

```python
print('Enter a number: ')
x = input()
val = len(x)
num = int(x)
output = calcArmstrong(num, val)
if output == num:
    print('It is an armstrong number')
else:
    print('It is not an armstrong number')
```

5. Write a program to print armstrong number in an interval

```python
print('Enter a number till where the armstrong number needs to be printed: ')
x = int(input())
for i in range(1, x+1):
    val = len(str(i))
    result = calcArmstrong(i, val)
    if result == i:
        print(i)
```

6. Write a program to find the sum of natural numbers

```python
def sumNatural(num):
    return num*(num+1)/2
```

```python
print('Enter a natural number: ')
num = int(input())
x = sumNatural(num)
print('Sum of natural numbers is: ', x)
```
