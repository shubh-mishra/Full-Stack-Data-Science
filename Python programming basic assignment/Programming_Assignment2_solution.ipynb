1. Write a python program to convert kilometers to miles

```python
def kmToMile(x):
    return x*0.621371
```

```python
print('Enter distance in kilometer: ')
x = float(input())
y = kmToMile(x)
print('Distance in miles is: ', y)
```

2. Write a python program to convert celsius to fahrenheit

```python
def celsiusToFahrenheit(x):
    return x*(9/5)+32
```

```python
print('Enter temperature in celsius: ')
x = float(input())
y = celsiusToFahrenheit(x)
print('Temperature in fahrenheit is: ', y)
```

4. Write a python program to solve quadratic equation

```python
from math import sqrt
def qSolve(a, b, c):
    D = b*b - 4*a*c
    x1 = (-b - sqrt(D))/(2*a)
    x2 = (-b + sqrt(D))/(2*a)
    return x1, x2
```

```python
print('a*x^2 + b*x + c = 0')
```

```python
print('Enter the coefficient of x^2 (a):')
a = float(input())
print('Enter the coefficient of x (b):')
b = float(input())
print('Enter the value of constant (c):')
c = float(input())
x1, x2 = qSolve(a, b, c) 
print('Solution: \n x1 = ', x1 ,'\n x2 = ', x2)
```

5. Write a python program to swap two variables without temp variable

```python
def swapVar(x, y):
    x = x+y
    y = x-y
    x = x-y
    print('x = ', x)
    print('y = ', y)
```

```python
print('Enter 1st variable(x): ')
x = int(input())
print('Enter 2nd variable(y): ')
y = int(input())
swapVar(x, y)
```
