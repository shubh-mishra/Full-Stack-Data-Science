2. Write a Python program to do arithmetic operations addition and division

Solution 2

```python
print('Enter 1st number(x): ')
x = int(input())
print('Enter 2nd number(y): ')
y = int(input())
print('x + y = ', x+y)
print('x / y = ', x/y)
```

3. Write a program to print area of triangle

Solution 3

```python
print('Enter the base of triangle: ')
b = int(input())
print('Enter the height of triangle: ')
h = int(input())
print('Area of triangle is: ', (b*h)/2)
```

4. Write a program to swap two variables

Solution 4

```python
print('Enter 1st variable(x): ')
x = input()
print('Enter 2nd variable(y): ')
y = input()
temp = x
x = y
y = temp
print('x = ', x)
print('y = ', y)
```
