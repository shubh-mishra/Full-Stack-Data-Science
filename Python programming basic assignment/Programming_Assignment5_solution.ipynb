1. Write a program to find LCM

```python
def lcm(n1, n2):
    if n1 > n2:
        greater = n1
    else:
        greater = n2
    while(True):
        if((greater % n1 == 0) and (greater % n2 == 0)):
            lcm = greater
            break
        greater += 1
    return lcm

num1 = 44
num2 = 24

print("The L.C.M. is", lcm(num1, num2))
```

2. Write a program to find HCF

```python
def hcf(n1, n2):
    if n1 > n2:
        smaller = n2
    else:
        smaller = n1
    for i in range(1, smaller+1):
        if((n1 % i == 0) and (n2 % i == 0)):
            hcf = i 
    return hcf

num1 = 44 
num2 = 24

print("The H.C.F. is", hcf(num1, num2))
```

3. Write a program to convert decimal to binary, octal, hexadecimal

```python
dec = 516
print("The decimal value of", dec, "is:")
print(bin(dec), "in binary.")
print(oct(dec), "in octal.")
print(hex(dec), "in hexadecimal.")
```

5. Write a program to make a simple calculator with 4 basic mathematical operations

```python
def add(a,b):
    return a+b
def subtract(a,b):
    return a-b
def multiply(a,b):
    return a*b
def divide(a,b):
    return a/b
```

```python
print('Select the operation: \n 1. Addition\n 2. Subtraction\n 3. Multiplication\n 4. Division (1,2,3,4): ')
x = int(input())
print('Enter the 1st operand (a): ')
a = int(input())
print('Enter the 2nd operand (b): ')
b = int(input())
if x == 1:
    y = add(a,b)
elif x == 2:
    y = subtract(a,b)
elif x == 3:
    y = multiply(a,b)
elif x == 4:
    y = divide(a,b)
else:
    print('Wrong selection')
print('The result is: ', y)
```
